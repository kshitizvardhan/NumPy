# Advance Operation with Business Examples

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# DataSet: [restaurant_id , year_2021, year_2022, year_2023, year_2024]

sales_data = np.array([
    [1, 10000, 12000, 15000, 18000],    # Chole Bhature Shots
    [2, 8000, 9000, 11000, 13000],      # Samosa Plaza
    [3, 15000, 16000, 17000, 20000],    # Sweets House
    [4, 5000, 7000, 9000, 12000],       # Vada Pav House
    [5, 20000, 22000, 25000, 30000]     # Biryani House
])
```

```python
print("====== Sales Analysis for 2021-2024 ======")
print("Shape of sales data:", sales_data.shape)
# total sales for each restaurant
total_sales = np.sum(sales_data[:, 1:], axis=1)
print("Total sales for each restaurant:", total_sales)
# total sales for each year
yearly_sales = np.sum(sales_data[:, 1:], axis=0)
print
# average sales per year for each restaurant
average_sales = np.mean(sales_data[:, 1:], axis=1)
print("Average sales per year for each restaurant:", average_sales)
# average sales per restaurant for each year
average_yearly_sales = np.mean(sales_data[:, 1:], axis=0)
print("Average sales per restaurant for each year:", average_yearly_sales)
# restaurant with highest total sales
top_restaurant_index = np.argmax(total_sales)
print("Restaurant ID with highest total sales:", sales_data[top_restaurant_index, 0])

cumsum = np.cumsum(sales_data[:, 1:], axis=1)
print("Cumulative sum of sales for each restaurant:\n", cumsum)

plt.figure(figsize=(10, 10))
plt.plot(np.mean(cumsum,axis=0))
plt.title('Average Cumulative Sales Over Years')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.grid(True)
plt.show()
```
