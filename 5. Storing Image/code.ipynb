```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
arr1 = np.array([1, 2, 3, 4, 5])
arr2 = np.random.rand(3, 3)
arr3 = np.zeros((4,4))
```

```python
np.save('array1.npy', arr1)
np.save('array2.npy', arr2)
np.save('array3.npy', arr3)
```

```python
load_arr1 = np.load('array1.npy')
print("Loaded Array 1:", load_arr1)
```

```python
try:
    # img = Image.open('logo.png')
    # logo_array = np.array(img)
    # np.save('logo.npy', logo_array)
    # print("Logo image saved as 'logo.npy' successfully.")
    logo = np.load('numpy-logo.npy')

    # display the image
    plt.figure(figsize=(8,8))
    plt.subplot(121)
    plt.imshow(logo)
    plt.title("NumPy Logo")
    plt.grid(False)
    plt.axis('off')  # Hide axes for better visualization
    plt.show()

    # CONVERTING TO DARK MODE
    dark_logo = 1 - logo  # Invert the colors for dark mode
    plt.figure(figsize=(8,8))
    plt.subplot(122)
    plt.imshow(dark_logo)
    plt.title("Dark Mode NumPy Logo")
    plt.grid(False)
    plt.axis('off')  # Hide axes for better visualization
    plt.show()

except FileNotFoundError:
    print("File not found. Please check the file path and try again.")
```
